#Importing the required libraries :

```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
```

# Reading and importing the dataset

```python
df=pd.read_csv("http://bit.ly/w-data")
print("Data imported successfully")
df.head(10)
```

# Plotting graph for analysing given data

```python
df.plot(x="Hours", y="Scores", style="o")
plt.xlabel("No. of study Hours")
plt.ylabel("Percentage")
plt.title("No. of study Hours vs Percentage")
plt.grid("True",color="#99ccff",linestyle="-.")
```

# Interpretation from graph :

From the above obtained graph, it is clearly seen that there is a **positive relation** between the number of study hours and the percentage of student


# Preparing the Data

Sorting the dependent and the independent variables of the dataset

```python
X=df.iloc[:, :-1].values
y=df.iloc[:, 1].values
```

# Splitting the data 

split the data into 2 parts - 'training set' and 'test set'

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1)
```

# Fitting the Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression().fit(X_train,y_train)
```

# Plotting the regression line and test data

```python
line = reg.coef_*X+reg.intercept_
plt.scatter(X, y)
plt.plot(X, line);
plt.show()
```


# Making Predictions

Predicting the scores

```python
y_pred = reg.predict(X_test)
```


Comparing Actual vs Predicted scores

```python
df1 = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
print(df1)
```

# Predicting score for given 9.25 hour/day

```python
new_pred = reg.predict([[9.25]])
print("Number of Hours: 9.25")
print("Predicted Score: ",new_pred)
```

# Calculating the MSE of the model

```python
np.mean((reg.predict(X_test) - y_test)**2)
```
